```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
```

```python
a=np.array([1,2,4,5])
```

```python
b=a**3
```

```python
b
```

```python
x=np.linspace(0,2*pi,100)
```

```python
y=np.sin(x)
z=np.cos(x)
```

```python
x
```

```python
plt.xlabel("Angles")
plt.ylabel("Sin(x)")
plt.title("Sin graph")
plt.plot(x,y,color="green")
```

```python
plt.plot(x,y,color="green",label="Sin(x)")
plt.plot(x,z,color="blue",label="Cos(x)")
plt.legend()
```

```python
plt.plot(x,y,'r-.',x,z,'g')
```
